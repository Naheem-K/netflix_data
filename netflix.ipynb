```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('netflix_titles.csv')
df = pd.DataFrame(data)
df.head(30)
```

```python

df[['rating','release_year']]
```

Trendings by year

```python

subset = df.groupby('release_year')['rating'].count()

plt.figure(figsize=(12,6))
plt.bar(subset.index, subset.values)
plt.title('Understanding Trends by Year', fontdict=dict(fontsize=15, family='Arial', fontweight='bold', color='black'))
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.dropna().head(10)
```

Trending categories by country

```python
st = df['country'].value_counts().head(10)

plt.pie(st.values,labels=st.index,
        autopct='%1.1f%%',
        startangle=170,
        )
plt.title('Trendings By Country',size=16,family='Arial')
plt.show()
```

relationship bitween duration and trendings


```python
subset = df.head(10)
year_counts = subset['duration'].value_counts().sort_index()


plt.figure(figsize=(10,6))
plt.plot(year_counts.index, year_counts.values, marker='o', color='green', linewidth=2)

plt.title('Content Release Trend Over Years', fontsize=16, family='Arial', color='darkblue')
plt.xlabel('Duration', fontsize=12)
plt.ylabel('rating', fontsize=12)
plt.grid(True, linestyle='--')
plt.show()
```
